```python
# Heatmap for Correlations
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),annot=True, cmap="coolwarm",fmt =".2f")
plt.title("Correlation Heatmap")
plt.show()
```

🔍 Key Takeaways from the Heatmap
All off-diagonal correlations are near 0: This means variables like TSH_Level, T3_Level, T4_Level, Nodule_Size, and Risk_Score are not linearly related to one another.

Diagonal values are 1.00: That’s expected—each variable is perfectly correlated with itself.

Color gradient confirms it visually: The heatmap uses a blue-to-red scale, but everything stays neutral (no deep red or blue), reinforcing the lack of correlation.



🧠 What This Means for Your Analysis
No multicollinearity: You won’t run into issues with redundant features in regression models.

Linear models may miss patterns: Since relationships aren’t linear, models like logistic regression might underperform unless you engineer features or use nonlinear methods.

Tree-based models will shine: Algorithms like Random Forest, XGBoost, or LightGBM can uncover nonlinear dependencies and interactions that correlation can’t detect.

Feature importance ≠ correlation: Just because T3_Level and Risk_Score aren’t correlated doesn’t mean T3 isn’t predictive. SHAP values or permutation importance might reveal hidden influence

```python
coun = df.groupby('Country')['Gender'].value_counts(normalize = True).unstack().reset_index()
for index, row in coun.iterrows():
    # Get the gender columns (assumes first column is 'Country')
    gender_labels = row.index[1:]
    gender_values = row[gender_labels].values
    
    plt.figure(figsize=(6,6))
    plt.pie(gender_values, labels=gender_labels, autopct='%1.1f%%', startangle=90)
    plt.title(f"Gender Distribution in {row['Country']}")
    plt.axis('equal') 
    plt.show()
```
